```python
                    #Problem 03(Compound Interest)
def compound_interest_hredoy(P, R, T):
    A = P * (1 + R/100) ** T
    return A

P = float(input("Principal Amount: "))
R = float(input("Interest Rate: "))
T = float(input("Time (years): "))

print("Compound Interest Amount =", compound_interest_hredoy(P, R, T))
```

```python
                  #Problem 04(Sum of Series)
N = int(input("Enter N: "))

total = 0

for i in range(1, N + 1):
    total += i ** 2

print("Sum =", total)
```

```python
                     #Problem  05(Prime Number Check)
def prime_finder(n):
    if n < 2:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

n = int(input("Enter a number: "))

if prime_finder(n):
    print("Prime Number")
else:
    print("Not Prime Number")
```

```python
                    #Problem 06 (Fibonacci Number)
n = int(input("Enter n: "))

a = 0
b = 1

if n == 0:
    print(a)
elif n == 1:
    print(b)
else:
    for i in range(2, n + 1):
        c = a + b
        a = b
        b = c

    print("Fibonacci Number =", b)
```

```python
                        #Problem 09(Largest and Smallest Element)
def largest_number_hredoy(lst):
    largest = lst[0]

    for num in lst:
        if num > largest:
            largest = num

    return largest


def smallest_number_hredoy(lst):
    smallest = lst[0]

    for num in lst:
        if num < smallest:
            smallest = num

    return smallest


numbers = [12, 45, 2, 89, 6, 100]

print("Largest =", largest_number_hredoy(numbers))
print("Smallest =", smallest_number_hredoy(numbers))
```

```python
                   #Problem 10(Second Largest Element)
numbers = [10, 50, 30, 70, 90, 80]

largest = second = -999999

for num in numbers:
    if num > largest:
        second = largest
        largest = num
    elif num > second and num != largest:
        second = num

print("Second Largest =", second)
```

```python
                        #Problem 13(Count "CSE303" in a String)
text = input("Enter a string: ")

count = 0
target = "CSE303"

for i in range(len(text) - len(target) + 1):
    match = True

    for j in range(len(target)):
        if text[i + j] != target[j]:
            match = False
            break

    if match:
        count += 1

print("Count =", count)
```

```python
                       #Problem 14(Palindrome Checker)
def palindrome_checker_hredoy(text):
    reverse = ""

    for i in range(len(text)-1, -1, -1):
        reverse += text[i]

    if text == reverse:
        return True
    else:
        return False


text = input("Enter a string: ")

if palindrome_checker_hredoy(text):
    print("Palindrome")
else:
    print("Not Palindrome")
```
