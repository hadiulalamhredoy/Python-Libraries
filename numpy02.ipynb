                            #Iterating Over Array  


```python
a= np.arange(0,45,5)
a= a.reshape(3,3)
a
```

```python
for x  in np.nditer(a):
    print(x)
```

                            #Iteration  Order C & F style

```python
print(a) 
for x in np.nditer(a, order="C"):
           print(x)

for x in np.nditer(a, order='F'):
           print(x)
```
